```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(2)
from sklearn.decomposition import PCA
```

```python
standardized_iris = pd.read_pickle(filepath_or_buffer='./pkls/dataset/standardized_iris')
print("Dataset shape:", standardized_iris.shape)
```

```python
pca = PCA(n_components=2)
pca.fit_transform(standardized_iris)
pca_projected_iris = pca.transform(standardized_iris)
display(pca.components_) # Each row is a pc
```

```python
pca_projected_iris = pd.DataFrame(pca_projected_iris)
```

```python
pd.to_pickle(pca_projected_iris, filepath_or_buffer='./pkls/projected_points/pca_sklearn_projected_points')
```

```python
plt.clf()
plt.scatter(
    x=pca_projected_iris.iloc[:,:1],
    y=pca_projected_iris.iloc[:,1:2],
    edgecolors='k',
    alpha=0.8
    )
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

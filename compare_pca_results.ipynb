```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
pca_eigen_decomposition_projected_points = pd.read_pickle(filepath_or_buffer='./pkls/projected_points/pca_eigen_decomposition_projected_points')
print("pca_eigen_decomposition_projected_points shape:", pca_eigen_decomposition_projected_points.shape)

pca_svd_projected_points = pd.read_pickle(filepath_or_buffer='./pkls/projected_points/pca_svd_projected_points')
print("pca_svd_projected_points shape:", pca_svd_projected_points.shape)

pca_sklearn_projected_points = pd.read_pickle(filepath_or_buffer='./pkls/projected_points/pca_sklearn_projected_points')
print("pca_sklearn_projected_points shape:", pca_sklearn_projected_points.shape)
```

```python

plt.clf()
fig = plt.figure()
fig.suptitle('Comparison of PCA Methods', fontsize=12)  

# Define the grid layout
gridSpec = GridSpec(2, 1, height_ratios=[1, 4], figure=fig)  


text_ax = fig.add_subplot(gridSpec[0])
text_content = """
Scikit-learn picked the same eigenvector in the data's direction but flipped its orientation,\n
causing the PCA to look mirrored without affecting the variance or the data's meaning.
"""
text_ax.text(0, 0.5, text_content, ha='left', va='center', fontsize=8)
text_ax.axis('off')  # Remove axes for the text box

scatter_plot = fig.add_subplot(gridSpec[1])
scatter_plot.scatter(
    x=pca_eigen_decomposition_projected_points.iloc[:, 0],
    y=pca_eigen_decomposition_projected_points.iloc[:, 1],
    c='red', 
    s=100,     
    label='Eigen Decomposition'
)
scatter_plot.scatter(
    x=pca_svd_projected_points.iloc[:, 0],
    y=pca_svd_projected_points.iloc[:, 1],
    c='blue', 
    s=50,     
    label='SVD'
)
scatter_plot.scatter(
    x=pca_sklearn_projected_points.iloc[:, 0],
    y=pca_sklearn_projected_points.iloc[:, 1],
    c='orange',   
    s=25,      
    label='SKLearn PCA'
)


scatter_plot.set_xlabel('PC1')
scatter_plot.set_ylabel('PC2')
scatter_plot.legend()

plt.tight_layout(rect=[0, 0, 1, 0.95]) 
plt.show()
```
